```python
import time
```

```python
print ('Informe os valores de a, b e c quando (ax^2+bx+c=0)')
```

```python
while True:

    a = int (input ('a: '))
    b = int (input ('b: '))
    c = int (input ('c: '))
    
    print (f'Equação: {a}x^2 + {b}x + {c} = 0')
    time.sleep (2)
    
    
    D = b**2 - 4 * a * c
    print (f'Delta = {D:.1f}')
    time.sleep (2)
    
      
    x1 = (-b + D**(1/2)) / (2 * a)
    x2 = (-b - D**(1/2)) / (2 * a)
    
    print (f'x1 = {x1:.1f}')
    print (f'x2 = {x2:.1f}')
    time.sleep (2)
    
    
    final = (input('Digite: \n "1" para continuar \n "2" para finalizar'))
    if final == '2': 
        break
```
